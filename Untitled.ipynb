```python
import mediapipe as me
import cv2
import numpy as np
import os
import time
import math
import pyautogui
import ctypes
from mediapipe.python.solutions.drawing_utils import _normalized_to_pixel_coordinates
from comtypes import CLSCTX_ALL
from pycaw.pycaw import AudioUtilities, IAudioEndpointVolume
import win32api
import win32con
import mouse

devices = AudioUtilities.GetSpeakers()
interface = devices.Activate(
    IAudioEndpointVolume._iid_, CLSCTX_ALL, None)
volume = interface.QueryInterface(IAudioEndpointVolume)
min_,max_,_=volume.GetVolumeRange()
```

```python
cap=cv2.VideoCapture(0)
prev_frame_time = 0
new_frame_time = 0

with hand.Hands(min_detection_confidence=0.8,min_tracking_confidence=0.4) as hands:
    while cap.isOpened():
        ret,frame=cap.read()
        
        image=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
        image=cv2.flip(image,1)
        
        image.flags.writeable=False
        result=hands.process(image)
        image.flags.writeable=True
        
        image=cv2.cvtColor(image,cv2.COLOR_RGB2BGR)
        
        if result.multi_hand_landmarks:
            for num,hand_cord in enumerate(result.multi_hand_landmarks):
                draw.draw_landmarks(image,hand_cord,hand.HAND_CONNECTIONS)
                
        
        new_frame_time = time.time()
        fps = 1/(new_frame_time-prev_frame_time)
        prev_frame_time = new_frame_time
        fps=str(fps)
        fps=fps[:4]
        cv2.putText(image, "FPS-"+fps, (7, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 100), 1, cv2.LINE_AA)

        
        cv2.imshow("Hand Track",image)
        if cv2.waitKey(1) & 0xFF ==ord("q"):
            break
    cap.release()
    cv2.destroyAllWindows()
```

```python
def volumne_(co,ploc_d):
    smooth_=12
    H8=_normalized_to_pixel_coordinates(co.landmark[8].x,co.landmark[8].y,image.shape[1],image.shape[0])
    H4=_normalized_to_pixel_coordinates(co.landmark[4].x,co.landmark[4].y,image.shape[1],image.shape[0])
    if H8==None or H4==None:
        return -1
    vol_range=int(abs(H8[0]-H4[0])+abs(H8[1]-H4[1]))

    if vol_range<=40:
        hypotenuse=np.sqrt((H8[0]-H4[0])**2+(H8[1]-H4[1])**2)
        opp=np.sqrt(((40+H4[0])-H8[0])**2+(H8[1]-H4[1])**2)
        deg=round(math.sin(opp/hypotenuse),2)
        if ploc_d!=None:
            clo_d=ploc_d+(deg-ploc_d)/smooth_

        return clo_d
    return -1
```

```python
def hand_up_down(co):
    H8=_normalized_to_pixel_coordinates(co.landmark[8].x,co.landmark[8].y,image.shape[1],image.shape[0])
    H5=_normalized_to_pixel_coordinates(co.landmark[5].x,co.landmark[5].y,image.shape[1],image.shape[0])
    if H8==None or H5==None:
        return 0
    return int(((H8[0]-H5[0])**2+(H8[1]-H5[1])**2))
    
```

```python
def right_distance_click(hand_cord):
    cord8 = _normalized_to_pixel_coordinates(hand_cord.landmark[8].x,hand_cord.landmark[8].y,image.shape[1],image.shape[0])
    cord12 = _normalized_to_pixel_coordinates(hand_cord.landmark[12].x,hand_cord.landmark[12].y,image.shape[1],image.shape[0])
    if  cord8!=None or cord12!=None:
        return np.sqrt((cord12[1]-cord8[1])**2+(cord12[0]-cord8[0])**2),cord8
    return -1
```

```python
def left_distance_click(hand_cord):
    cord4 = _normalized_to_pixel_coordinates(hand_cord.landmark[4].x,hand_cord.landmark[4].y,image.shape[1],image.shape[0])
    cord12 = _normalized_to_pixel_coordinates(hand_cord.landmark[12].x,hand_cord.landmark[12].y,image.shape[1],image.shape[0])
    if  cord4!=None or cord12!=None:
        return np.sqrt((cord12[1]-cord4[1])**2+(cord12[0]-cord4[0])**2)
    return -1
```

```python
cap=cv2.VideoCapture(0)
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 900)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT,500)
smooth=5
plocx,plocy=0,0
clox,cloy=0,0
ploc_d,clo_d=0,0
volu=0
p_vol=0
c=0
with hand.Hands(min_detection_confidence=0.8,min_tracking_confidence=0.4) as hands:
    while cap.isOpened():
        ret,frame=cap.read()   
        
        image=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
        image=cv2.flip(image,1)
        
        image.flags.writeable=False
        result=hands.process(image)
        image.flags.writeable=True
        
        image=cv2.cvtColor(image,cv2.COLOR_RGB2BGR)
        
        if result.multi_hand_landmarks:
            for num,hand_cord in enumerate(result.multi_hand_landmarks):
                if result.multi_handedness[0].classification[0].label=="Right":
                    draw.draw_landmarks(image,hand_cord,hand.HAND_CONNECTIONS)
                    
                    
                    
#--------------------------------------------------------------------------------------------------------------        
                    #Volumne 
                    p_vol=np.interp(volume.GetMasterVolumeLevel(),[min_,max_],[0,1])
                    volu=volumne_(hand_cord,p_vol)
                    if volu!=-1 and volu!=None:
                        vo=int(np.interp(volu,[0,1],[min_,max_]))
                        show_vol=int(np.interp(vo,[min_,max_],[500,120]))
                        percent=int(np.interp(vo,[min_,max_],[0,100]))
                        cv2.putText(image,str(percent)+"%",(70,110),cv2.FONT_HERSHEY_COMPLEX,2,(0,0,255),3)
                        cv2.rectangle(image,(79,120),(120,500), (255,0,0),3)
                        cv2.rectangle(image,(80,show_vol),(120,500), (0,0,255),-1)
                        volume.SetMasterVolumeLevel(vo, None)
                        time.sleep(0.05)
                        
                    
                    
                    # Mouse move
                    
                    else:
                            #cv2.rectangle(image, (470,300), (600,400), (0,255,255), 4) 
                            dist_right,cord8=right_distance_click(hand_cord)
                            dist_left=left_distance_click(hand_cord)
                            if dist_right>34 and dist_left>20:
                                xs=int(np.interp(cord8[0],(470,600),(0,SW)))
                                ys=int(np.interp(cord8[1],(300,400),(0,SH)))
                                clox=int(plocx+(xs-plocx)/smooth)
                                cloy=int(plocy+(ys-plocy)/smooth)
                                ctypes.windll.user32.SetCursorPos(clox,cloy)
                                plocx,plocy=clox,cloy
                                ctypes.windll.user32.SetCursorPos(clox,cloy)
                            
                            #Right_click
                            elif dist_right<34 and dist_right!=-1:
                                mouse.right_click()
                                time.sleep(0.4)
                            
                            #Left _click
                            elif dist_left<20 and dist_left!=-1:
                                mouse.press(button='left')
                                mouse.release(button='left')
                                time.sleep(0.4)
                    
                                    
                    
#--------------------------------------------------------------------------------------------------------------        

        cv2.imshow("Hand Track",image)
        if cv2.waitKey(1) & 0xFF ==ord("q"):
            break
    cap.release()
    cv2.destroyAllWindows()
```
